```python


# turning into a function
def find_(predict_list, label_list):
    # creates tuple
    # correct preds is ordered: index, prediction
    # incorrectpreds is ordered: index, prediction, true label
    correctPreds = []
    incorrectPreds= []
    predict = ast.literal_eval(predict_list)
    label = ast.literal_eval(label_list)
    
    for ndx, n in enumerate(predict):
            if n == label[ndx]: # match correct prediction
                #print("ndx", ndx,"\n ", n )
                correctPreds.append((ndx, n))
            elif n != label[ndx]: # incorrect prediction
                incorrectPreds.append((ndx, n, label[ndx]))
                
    return correctPreds, incorrectPreds
```
